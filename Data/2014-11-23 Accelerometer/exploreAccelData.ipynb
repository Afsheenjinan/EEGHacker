### Purpose

Record accelerometer data from OpenBCI V3 board during known motions.  Confirm that the data matches the known motions.

### Setup

OpenBCI V3 Board (see below) with software as delivered from OpenBCI.  Data was logged using the OpenBCI GUI in Processing.

<img src="Pics/BoardWithBatteries.jpg" style="width: 480px;" align="left"/>

### Procedure 

I inserted the batteries to my OpenBCI board to give it power.  I started the OpneBCI GUI to begin recording data.  Holding the board in my hand, I completed the following manuevers:

1. Start with board flat and level (z-axis points up)
2. Roll it to the right (x-axis points down) and left (x-axis points up)
3. Tip it nose down (y-axis points down) and nose up (y-axis points up)
4. Flip it upside down (z-axis points down)

Notice the markings on the OpenBCI board that indicate the direction of the axes.
<img src="Pics/BoardZoom.jpg" style="width: 480px;" align="left"/>

### Data

The 3-axis accelerometer data was saved to a text file by the Processing GUI.  The acceleration values were saved in engineering units by the Processing GUI.  The units are believed to be millimeters per second squared.

### Data Analysis

The data are processed here in Python.  The goal is to confirm that the data is well behaved, that the units are correct, that the scale factors are correct, and that the correct axes are responding to the known motions.

### Results

```python
# import libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define Data to Load
pname = 'SavedData/'
fname = 'OpenBCI-RAW-2014-11-23_18-54-57.txt'

# load data into numpy array
fs_Hz = 250.0        # assumed sample rate for the EEG data
data = np.loadtxt(pname + fname,
                  delimiter=',',
                  skiprows=5)

# check the packet counter for dropped packets
data_indices = data[:, 0]   # the first column is the packet index
d_indices = data_indices[2:]-data_indices[1:-1]
n_jump = np.count_nonzero((d_indices != 1) & (d_indices != -255))
print("Number of discontinuities in the packet counter: " + str(n_jump))
```

```python
# parse the data
#eeg_data_uV = data[:, 1:(8+1)]   # ignore the first channel (column 0), so channel 1 is column 1
accel_data_G = data[:, 9:(11+1)]/1000.0/9.806  #acceleromater data...convert mm/s2 to G
t_sec = np.arange(len(accel_data_G[:, 0])) / fs_Hz

# the data is padded with lots of zeros.  Only get the non-zero entries
I = np.nonzero(accel_data_G[:,0])
I=I[0]
print("Accel data exists in " + str(len(I)) + " of " + str(len(t_sec)) + \
      " (" + str((100*len(I))/len(t_sec)) + "%) of data packets");
accel_data_G = accel_data_G[I,:]
t_sec = t_sec[I]
```

```python
# plot each channel of accelerometer data
for Iplot in range(3):      
    plt.figure(figsize=(9.0, 2.5))
    plt.plot(t_sec,accel_data_G[:, Iplot])
    plt.plot([t_sec[0], t_sec[-1]],[0, 0],'k--')
    plt.plot([t_sec[0], t_sec[-1]],[1.0, 1.0],'k:')
    plt.plot([t_sec[0], t_sec[-1]],[-1.0, -1.0],'k:')
    plt.xlabel("Time (sec)")       
    plt.ylabel("Acceleration (G)")
    plt.title("Accelerometer Channel " + str(Iplot))
    plt.ylim([-1.5, 1.5])
    if (Iplot==0) :
        plt.text(1.0,0.25,"Flat and Level",horizontalalignment='left',backgroundcolor='w',verticalalignment='center')
        plt.text(10.5,-0.8,"Roll Right",horizontalalignment='right',backgroundcolor='w',verticalalignment='center')
        plt.text(18.5,0.8,"Roll Left",horizontalalignment='left',backgroundcolor='w',verticalalignment='center')
    elif (Iplot==1) :
        plt.text(20,-0.8,"Nose Down",horizontalalignment='right',backgroundcolor='w',verticalalignment='center')
        plt.text(27,0.8,"Nose Up",horizontalalignment='left',backgroundcolor='w',verticalalignment='center')
    elif (Iplot==2) :
        plt.text(30.5,-0.8,"Upside Down",horizontalalignment='right',backgroundcolor='w',verticalalignment='center')
       
        
```

```python
# compute magnitude of the acceleration vector
mag_accel_G = np.sqrt(accel_data_G[:,0]**2 + accel_data_G[:,1]**2 + accel_data_G[:,2]**2)
median_mag_accel_G = np.median(mag_accel_G)

print "Median acceleration should be near 1.0 G"
print "Median acceleration is actually " + "{:.3f}".format(median_mag_accel_G) + " G"
```

```python
#plot
plt.figure(figsize=(6.0, 2.5))
plt.plot(t_sec,mag_accel_G)
plt.plot([t_sec[0], t_sec[-1]],[1.0, 1.0],'k--')
plt.xlabel("Time (sec)")       
plt.ylabel("Acceleration (G)")
plt.title("Magnitude of Acceleration")
plt.ylim([0.6, 1.4])
plt.tight_layout()
plt.text(1,1.3,"Median = " + "{:.3f}".format(median_mag_accel_G) + "G")
```
