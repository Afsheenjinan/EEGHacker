```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
from scipy import signal
```

### Load the Data

```python
# Define Data to Load...be sure to UNZIP the data!!!
pname = 'SavedData/'
test_id = 0;

t_lim_sec = [0.0, 0.0] #default "empty" value

plot_case = 152;
if (plot_case == 0):
    #auditory stimulation
    fname = 'OpenBCI-RAW-2015-01-25_18-28-58_eyesOpen_AMSine.txt';
    t_lim_sec = [110.0, 280.0]
    test_id = 1
elif (plot_case == 152):
    #day 1...5 people controlling the shark
    fname = 'OpenBCI-RAW-2015-02-08_12-23-49_Day2_5manShark.txt'
    chan_names = ["Chan 1", "Chan 2", "Chan 3", "Chan 4", "Chan 5"]
    
elif (plot_case == 203):
    #day 2...3 people controlling one hex bug...
    #   chan1 = Rodrigo, "Left"
    #   chan 2 = Chip, "Forward"
    #   chan 3 = Joel, "Right"
    fname = 'OpenBCI-RAW-2015-02-08_14-32-09_Day2_3manHexBug.txt';
    chan_names = ["Left (Rodrigo)", "Forward (Chip)", "Right (Joel)"]
    t_lim_sec = [375, 925]
    t_lim_sec = [650, 750]
elif (plot_case == 204):
    #day 2...3 people controlling one hex bug...
    #   chan1 = Rodrigo, "Left"
    #   chan 2 = Chip, "Forward"
    #   chan 3 = Joel, "Right"
    fname = 'OpenBCI-RAW-2015-02-08_14-51-21_Day2_3manHexBug.txt';
    chan_names = ["Left (Rodrigo)", "Forward (Chip)", "Right (Joel)"]
    t_lim_sec = [450, 670]

    

# load data into numpy array
fs_Hz = 250.0        # assumed sample rate for the EEG data
data = np.loadtxt(pname + fname,
                  delimiter=',',
                  skiprows=5)
```

### Check the Packet Counter

```python
# check the packet counter for dropped packets
data_indices = data[:, 0]   # the first column is the packet index
d_indices = data_indices[2:]-data_indices[1:-1]
n_jump = np.count_nonzero((d_indices != 1) & (d_indices != -255))
print("Number of discontinuities in the packet counter: " + str(n_jump))
```

### Unpack the Data

```python
# parse the data out of the values read from the text file
eeg_data_uV = data[:, 1:(len(chan_names)+1)] # EEG data, only keep some of the channels. Units = microvolts
#accel_data_counts = data[:, 9:(11+1)]  # note, accel data is NOT in engineering units

# convert units
#unused_bits = 4 #the 4 LSB are unused?
#accel_data_counts = accel_data_counts / (2**unused_bits) # strip off this extra LSB
#scale_G_per_count = 0.002  # for full-scale = +/- 4G, scale is 2 mG per count
#accel_data_G = scale_G_per_count*accel_data_counts # convert to G

# create a vector with the time of each sample
t_sec = np.arange(len(eeg_data_uV[:, 0])) / fs_Hz
```

### Plot the EEG Data

```python
# plot each channel of EEG data...raw, to see what happens with the DC
nchan = eeg_data_uV.shape[2-1]  #normally 8 or 16
#ncol = 2; 
#nrow = (nchan+1) / ncol  # integer division!  This is my way of doing ceil(nchan/ncol)
#ncol = 1
#nrow = nchan
plt.figure(figsize=(9, 5))
plt.plot(t_sec,eeg_data_uV)
plt.xlabel("Time (sec)")       
plt.ylabel("Raw EEG (uV)")
plt.title(fname)
plt.ylim(4.5/24*1000000*np.array([-1.0, 1.0]))
plt.legend(chan_names)
if (t_lim_sec[1] != 0):
    plt.xlim(t_lim_sec)
else:
    plt.xlim([t_sec[0], t_sec[-1]])

```

### Spectrograms

```python
NFFT = 256*2  # pick the length of the fft
FFTstep = 1.0*fs_Hz  # do a new FFT every certain number of points (do a half or a whole second)
overlap = NFFT - FFTstep  # compute how much overlap between each subsequent FFT
f_lim_Hz = [0, 25]   # frequency limits for plotting

plt.figure(figsize=(10, nrow*4))
for Ichan in range(nchan):      
    ax = plt.subplot(nrow,ncol,Ichan+1)
    data = np.array(eeg_data_uV[:,Ichan])
    data = data - np.mean(data,0) # helps remove DC bleed into the low-freq bins
    spec_PSDperHz, freqs, t = mlab.specgram(data,
                               NFFT=NFFT,
                               window=mlab.window_hanning,
                               Fs=fs_Hz,
                               noverlap=overlap
                               ) # returns PSD power per Hz
    spec_PSDperBin = spec_PSDperHz * fs_Hz / float(NFFT)  #convert to "per bin"

    plt.pcolor(t, freqs, 10*np.log10(spec_PSDperBin))  # dB re: 1 uV
    plt.clim(20+np.array([-40, 0]))
    if (t_lim_sec[1] != 0):
        plt.xlim(np.array(t_lim_sec)+np.array([-10, 10]))
    else:
        plt.xlim([t_sec[0], t_sec[-1]])
    plt.ylim(f_lim_Hz)
    plt.xlabel('Time (sec)')
    plt.ylabel('Frequency (Hz)')
    plt.title("Channel " + str(Ichan+1) + ": " + chan_names[Ichan])

    # add annotation for FFT Parameters
    ax.text(0.025, 0.95,
        "NFFT = " + str(NFFT) + "\nfs = " + str(int(fs_Hz)) + " Hz",
        transform=ax.transAxes,
        verticalalignment='top',
        horizontalalignment='left',
        backgroundcolor='w')
    plt.colorbar()

plt.tight_layout()
```

### Filter to Just Detect Alpha Waves

```python
# filter the data to remove DC
hp_cutoff_Hz = 1.0
b, a = signal.butter(2, hp_cutoff_Hz/(fs_Hz / 2.0), 'highpass')  # define the filter
f_eeg_data_uV = signal.lfilter(b, a, eeg_data_uV, 0) # apply along the zeroeth dimension

# notch filter the data to remove 60 Hz and 120 Hz
notch_freq_Hz = np.array([60.0, 120.0])  # these are the center frequencies
for freq_Hz in np.nditer(notch_freq_Hz):  # loop over each center freq
    bp_stop_Hz = freq_Hz + 3.0*np.array([-1, 1])  # set the stop band
    b, a = signal.butter(3, bp_stop_Hz/(fs_Hz / 2.0), 'bandstop')  # create the filter
    f_eeg_data_uV = signal.lfilter(b, a, f_eeg_data_uV, 0)  # apply along the zeroeth dimension


#calculate SNR by calculating in-band and out-of-band signal power
foo_f_eeg_data_uV = f_eeg_data_uV
for Ifilt in [0, 1]:
    if Ifilt == 0:
        bp_Hz = np.array([9.0, 13.0])
    else:
        bp_Hz = np.array([15.0, 25.0])
    
    
    #bandpass filter
    b, a = signal.butter(3, bp_Hz/(fs_Hz / 2.0), 'bandpass')
    f_eeg_data_uV = signal.lfilter(b, a, foo_f_eeg_data_uV, 0)

    #rectify and low-pass filter
    lp_Hz = 0.5
    b, a = signal.butter(3, lp_Hz/(fs_Hz / 2.0),'lowpass')
    data_pow_uV2 = f_eeg_data_uV**2
    data_pow_uV2 = signal.lfilter(b,a,data_pow_uV2,0)
    data_pow_uV2[data_pow_uV2 < 0.0001] = 0.0001  #remove negative values prior to sqrt
    f_ampl_data_uV = np.sqrt(data_pow_uV2)
    
    if (Ifilt == 0):
        sig_pow_uV2 = data_pow_uV2
        sig_ampl_data_uV = f_ampl_data_uV
    else:
        noise_pow_uV2 = data_pow_uV2;
        noise_ampl_data_uV = f_ampl_data_uV

SNR_dB = 10.0*np.log10(np.divide(sig_pow_uV2,noise_pow_uV2))


# plots
plt.figure(figsize=(9, 5))
plt.plot(t_sec,SNR_dB)
plt.xlabel("Time (sec)")       
plt.ylabel("Filtered EEG (uV)")
plt.title(fname)
plt.ylim([0, 15])
plt.legend(chan_names)
if (t_lim_sec[1] != 0):
    plt.xlim(t_lim_sec)
else:
    plt.xlim([t_sec[0], t_sec[-1]])
    
plt.plot([t_sec[0], t_sec[-1]], 7.5*np.array([1.0, 1.0]),'k--',linewidth=2)
#plt.xlim([700, 710])
```
